```python
#  Регулярные выражения: Поиск и обработка текста
# Авторы: Шлегель Матвей, Мамедов Рамин | Группа 2-ИС33

import re


data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Суммы: $150.50, 20 EUR, 500 RUB, £75.
"""

print("=" * 60)
print(" ПРАКТИЧЕСКАЯ РАБОТА №7: Регулярные выражения")
print("=" * 60)


print("\n📧 ЗАДАНИЕ 1: Извлечение email-адресов")
print("-" * 40)

# Паттерн: 1+ символов (буквы/цифры/./-/_), @, домен, точка, 2-3 буквы
email_pattern = r'[\w.-]+@[\w-]+\.[a-zA-Z]{2,3}'
emails = re.findall(email_pattern, data_text)

print(f"Паттерн: {email_pattern}")
print(f"Найдено email: {emails}")


print("\n ЗАДАНИЕ 2: Извлечение цифр из кодов ID_PROD- и ID-")
print("-" * 40)

# Паттерн с группой: (?:...) - незахватывающая группа для префикса, (\d+) - захват цифр
code_pattern = r'(?:ID_PROD-|ID-)(\d+)'
code_numbers = re.findall(code_pattern, data_text)

print(f"Паттерн: {code_pattern}")
print(f"Исходные коды в тексте: ID_PROD-1001, ID-300-B")
print(f"Извлечённые числа: {code_numbers}")


print("\n ЗАДАНИЕ 3: Замена формата даты")
print("-" * 40)

# Паттерн с тремя группами: день, месяц, год
date_pattern = r'(\d{2})\.(\d{2})\.(\d{4})'
# Замена: \3 (год) - \2 (месяц) - \1 (день)
date_replacement = r'\3-\2-\1'

# Показываем до и после
print(f"До замены: 12.02.2025")
new_text = re.sub(date_pattern, date_replacement, data_text)
found_dates = re.findall(date_pattern, data_text)
print(f"Найдено дат в формате ДД.ММ.ГГГГ: {found_dates}")
print(f"После замены (фрагмент): ...{new_text.split('Коды')[0].strip()}...")

print("\n ЗАДАНИЕ 4: Суммы в EUR и RUB")
print("-" * 40)

# Паттерн: число (возможно с десятичной частью), пробел(ы), валюта
currency_pattern = r'\d+(?:\.\d+)?\s*(?:EUR|RUB)'
currency_values = re.findall(currency_pattern, data_text)

print(f"Паттерн: {currency_pattern}")
print(f"Найдено сумм: {currency_values}")


print("\n" + "=" * 60)
print(" РЕЗУЛЬТАТЫ ВСЕХ ЗАДАНИЙ (сводка)")
print("=" * 60)

def process_text_with_regex(text):
    """Обрабатывает текст всеми регулярными выражениями из заданий."""
    return {
        "emails": re.findall(r'[\w.-]+@[\w-]+\.[a-zA-Z]{2,3}', text),
        "product_codes": re.findall(r'(?:ID_PROD-|ID-)(\d+)', text),
        "dates_converted": re.sub(r'(\d{2})\.(\d{2})\.(\d{4})', r'\3-\2-\1', text),
        "eur_rub_sums": re.findall(r'\d+(?:\.\d+)?\s*(?:EUR|RUB)', text)
    }

results = process_text_with_regex(data_text)
print(f" Email: {results['emails']}")
print(f" Коды товаров (цифры): {results['product_codes']}")
print(f" Суммы в EUR/RUB: {results['eur_rub_sums']}")
print(f" Текст с заменёнными датами:\n{results['dates_converted']}")
```
